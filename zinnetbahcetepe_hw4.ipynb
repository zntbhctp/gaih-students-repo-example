```python
adiniz = input("Adınızı giriniz :")
print("merhaba " + adiniz + " Hangman oyununa hoşgeldiniz")

gizli_kelime = "snowhunter"
tahmininiz = ""

can = 10

while can > 0:
    harf_left = 0

    for karakter in gizli_kelime:
        if karakter in tahmininiz :
            print(karakter)
        else:
            print("_")
            harf_left += 1

    if harf_left == 0:
        print("Tebrikler kazandınız!")
        break

    ziyaretci = input("lütfen harf girin :")
    tahmininiz += ziyaretci

    if ziyaretci not in gizli_kelime:
        can -= 1
        print("hatalı !")
        print(f"kalan can hakkınız {can} ")

        if can == 0:
            print("canınız bitti")
```
